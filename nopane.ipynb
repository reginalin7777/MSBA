```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
nopane = pd.read_csv('Nopane.csv')
```

## Regression 1

```python
my_formula="UnitSales~1+NopaneAdDollars+CompetitionAdExpend+DumSegment+DumCopy"
reg1 = smf.ols(my_formula,data=nopane).fit()
print (reg1.summary())
```

### Checking model assumption

```python
stats.probplot(reg1.resid, dist="norm", plot=plt)
plt.show()

plt.scatter(reg1.fittedvalues,reg1.resid)
plt.axhline(y=0, color="red")
plt.title("Residual vs. fit plot")
plt.xlabel("fitted value")
plt.ylabel("residual")
```
