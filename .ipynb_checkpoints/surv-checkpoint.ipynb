# Notice that there is limited resource in doing survival analysis by Python
# So we just demo the KM Curve and basic cox regression fitting here

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv('HR.csv')
```

# KM Curve

* By Salary

```python
from lifelines import KaplanMeierFitter
km = KaplanMeierFitter()
#km.fit(data['time_spend_company'], event_observed = data['left'])
#km.plot()

#by salary 
salary = data["salary"]
dept = data['department']

first = True 
for sala in list(np.unique(salary)):
    d = data[data['salary']==sala]
    km.fit(d['time_spend_company'], event_observed = d['left'], label = sala)
    if first == True:
        ax = km.plot() 
    if first == False:
        km.plot(ax=ax)
    first = False
plt.title('KM curve by salary')
```

* By department

```python
dept_unique = list(np.unique(dept))
first = True 
for de in dept_unique:
    d = data[data['department']==de]
    km.fit(d['time_spend_company'], event_observed = d['left'], label = de)
    if first == True:
        ax = km.plot() 
    if first == False:
        km.plot(ax=ax)
    first = False
plt.title('KM curve by sales')
```

# Cox Regression Fit

```python
#Creating dummy vairbales for Categorical predictors, because sklearn can only handle numeric data
data['department']=data['department'].str.lower()
data=data.sort_values(by=['department','salary'])

data_dumm = pd.get_dummies(data[['department', 'salary']],drop_first=True)
data_dumm['time_spend_company'] = data['time_spend_company']
data_dumm['left'] = data['left']
data_dumm['satisfaction_level'] = data['satisfaction_level']
data_dumm['last_evaluation'] = data['last_evaluation']
data_dumm['number_project'] = data['number_project']
data_dumm['average_monthly_hours'] = data['average_montly_hours']
data_dumm['Work_accident'] = data['Work_accident']
data_dumm['promotion_last_5years'] = data['promotion_last_5years']
```

```python
from lifelines import CoxPHFitter 
cph = CoxPHFitter()
cph.fit(data_dumm, 'time_spend_company', event_col= 'left')
cph.print_summary()
```
